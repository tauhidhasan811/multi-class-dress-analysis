```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_train = np.load('data/X_train.npy')
X_val = np.load('data/X_val.npy')
y_train = np.load('data/y_train.npy')
y_val = np.load('data/y_val.npy')
```

```python
X_train[3].shape
```

```python
labels=[]
count=0
with open('data/split/list_attr_cloth.txt') as file : 
    i=0
    for f in file:
        i+=1
        label = f.strip()
        label = label.split(' ')
        idx = int(label[-1])
        if(len(labels) < idx):
            count+=1
            labels.append([label[0]])
        else:
            labels[idx-1].append(label[0])
labels
```

```python
length =[]
for i in range(len(labels)):
    length.append(len(labels[i]))
length
```

```python
idx = [random.randint(0, 5000) for i in range(30)]
count=0
plt.figure(figsize=(30, 10))
for i in idx:
    plt.subplot(3, 10, count+1)
    plt.imshow(X_train[i])
    plt.xticks([])
    plt.yticks([])
    plt.xlabel([ labels[i][v] for i, v in enumerate(y_train[1]) if i > 2])
    plt.ylabel([ labels[i][v] for i, v in enumerate(y_train[1]) if i < 3])
    count+=1
plt.show()
```
