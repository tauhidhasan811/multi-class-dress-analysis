```python
root = 'data'
```

```python
with open('data/split/train.txt') as file:
    train_path = [ os.path.join(root,p.strip()) for p in file]
len(train_path)
```

```python
with open('data/split/val.txt') as file:
    val_path = [ os.path.join(root,p.strip()) for p in file]
len(val_path)
```

```python
train_label = []
with open('data/split/train_attr.txt') as file:
    for l in file:
        l = l.strip()
        label = [int(la) for la in l.split(' ')]
        #label = [int(la) for la in label]
        train_label.append(label)
len(train_label)
```

```python
validate_label = []
with open('data/split/val_attr.txt') as file:
    for l in file:
        l = l.strip()
        label = [int(la) for la in l.split(' ')]
        #label = [int(la) for la in label]
        validate_label.append(label)
len(validate_label)
```

```python
train_data=[]
img_shape = (225, 300)
for path in train_path:
    img = cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, img_shape)
    train_data.append(img)
len(train_data)
train = np.array(train_data)
tra_label = np.array(train_label)
print(f'Data shaape : {train.shape}\nLabel Shape: {tra_label.shape}')
```

```python
val_data=[]
img_shape = (225, 300)
for path in val_path:
    img = cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, img_shape)
    val_data.append(img)
len(val_data)
validation = np.array(val_data)
val_label = np.array(validate_label)
print(f'Data shaape : {validation.shape}\nLabel Shape: {val_label.shape}')
```

```python
np.save(f'{root}/X_train.npy', train_data)
np.save(f'{root}/X_val.npy', validation)
np.save(f'{root}/y_train.npy', tra_label)
np.save(f'{root}/y_val.npy', val_label)
```
